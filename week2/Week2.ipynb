# 4. Methods

&gt; Methods are functions defined in a class

&gt; Methods usually define the behavior of the class

&gt; Methods just like functions can take multiple arguments

&gt; Methods should always define an argument called ‘self’

&gt; Self in python classes refer to the instance or the object created from the class

```python

     #Method logic
```

````python
#Example

#Create class dog, this can be used to create multiple dog objects
class Dog:
    'This class represents a dog'
    #Class variable or attribute name to hold the dogs breed
    breed = 'Rocky'

    #Creating a method to make the dog Bark( behavior )
    def bark(self):
        print('Bow Wow!!')

#Create an instance of the dog and call it dog1, dog1 is a new object created from the dog class
dog1 = Dog()

#Get the description of the class dog and print it
dog1_class_description = dog1.__doc__
print(dog1_class_description)

#Get the attribute breed of dog1 and print it
dog1_breed = dog1.breed
print('Dog1\'s breed is : ' + dog1_breed)

#Make the dog bark
dog1.bark();
````

&gt; Constructor is a method which Python calls whenever an instance of a class is created

**Usage**
```python
#Create a method __init__ within the class

  #Constructor logic

  ```



````python
#Example

#Create class dog, this can be used to create multiple dog objects
class Dog:
    'This class represents a dog'
    #Class variable or attribute name to hold the dogs name
    breed = 'Bulldog'

    #Creating a constructor which will be called everytime a new object is created
    def __init__(self):
        print(self.breed)
        print('Dog Object created')

#Create an object dog1 from dog
dog1 = Dog()

#Create an object dog2 from dog
dog2 = Dog()

dog3 = Dog()
````

# 6. Class Variables

&gt; Class variables are attributes or variables which will be shared by all instances of the class

&gt; If the class variable is updated from one instance or one single object of a class, it gets updated for all existing instances of the same class

&gt; Class variables are used when an attribute has to be shared across all instances of the class

&gt; Class variables are always defined in the class block of the code

&gt; In our previous examples the  attribute ```'breed.variable'``` is a class variable since it is shared by ```dog1``` and ```dog2```

&gt; Class variables can be accesed using the ``` className.variableName``` instead of instance Name ( Example ```dog.breed``` )

&gt; Class variables can be updated using ```className.variableName = 'New Value'```


````python
#Example

class Dog :
   #Class variable
   breed = 'Poodle'

#Accesed via the class Name
print(Dog.breed)

#Create new instance dog1
dog1 = Dog()

#Access variable breed via dog1 instance name
print(dog1.breed)

#Create new instance dog2
dog2 = Dog()

#Access variable breed via dog2 instance name
print(dog2.breed)

Dog.breed = 'Bulldog'

dog3 = Dog()
print(dog3.breed)
print(dog2.breed)
````

# 7. Inheritance and Polymorphism

# Inheritance

&gt; Object-oriented programming allows classes to inherit commonly used state and behavior from other classes

&gt; This technique is called inheritance

&gt; In inheritance there is usually a parent class from which multiple child classes can inherit

&gt; The child classes will get almost all of the features from the parent class

&gt; Lets consider Animal is a parent class
&gt; Animal has attributes : color, breed, name, isPet etc..
&gt; Animal has methods : eat(), sleep(), walk(), makeNoise()

&gt; Lets consider a child class Dog which is inherited from Animal
&gt; Dog will have the same attributes and methods as the class Animal, without them being explicitly defined in Dog class

&gt; Lets consider another child class Cat which is inherited from Animal
&gt; Cat will have the same attributes and methods as the class Animal, without them being explicitly defined in Dog class


**Usage**

```python
class ParentClass()
    init;
    attributes;
    methods;

class childClass(ParentClass)
    init;

```

````python
#Example

#Create an Animal Class
class Animal():
    'This is a Parent class for creating Animals'

    def __init__(self,name,breed,color):
        self.name = name
        self.breed = breed
        self.color = color


    def walk(self):
        print(self.name + ' is walking')

    def eat(self):
        print(self.name + ' is eating')

    def makeNoise(self):
        print(self.name + ' is making noise')

class Dog(Animal):
    'This is the Dog class inherited from Animal'

    '''The keyword pass is used when you do not have any statements or expressions.
      In this case Dog will inherit everything from the animal class so we dont need any code here, just the class'''

    pass

class Cat(Animal):
    'This is the cat class inherited from Animal'
    pass


#Create a new Dog object dog1
dog1 = Dog('max','bulldog','black')

#Create a new Cat object cat1
cat1 = Cat('manny','tabby','orange')

dog1.eat()
cat1.eat()
dog1.makeNoise()
cat1.makeNoise()
````

# Method overriding:

&gt; Changing the functionality of a method in a Child Class inherited from a Parent class is called overriding

&gt; Example : Animal Class has a method makeNoise(). Dogs bark and cats meow.

&gt; makeNoise() can be overridden in the Dog class to make the dog bark and in the cat class to make the cat meow


````python
#Example ( same as the one from last slide but updated Dog and Cat classes )

#Create an Animal Class
class Animal():
    'This is a Parent class for creating Animals'

    def __init__(self,name,breed,color):
        self.name = name
        self.breed = breed
        self.color = color


    def walk(self):
        print(self.name + ' is walking')

    def eat(self):
        print(self.name + ' is eating')

    def makeNoise(self):
        print(self.name + ' is making noise')

class Dog(Animal):
    'This is the Dog class inherited from Animal'

    #Override the makeNoise method to make the dog bark
    def makeNoise(self):
        print(self.name + ' is barking')

class Cat(Animal):
    'This is the cat class inherited from Animal'

    # Override the makeNoise method to make the cat meow
    def makeNoise(self):
        print(self.name + ' is meowing')


#Create a new Dog object dog1
dog1 = Dog('max','bulldog','black')

#Create a new Cat object cat1
cat1 = Cat('manny','tabby','orange')

dog1.eat()
cat1.eat()
dog1.makeNoise()
cat1.makeNoise()
````

# Multiple Inheritance

&gt; A class can inherit from multiple classes

&gt; This is called Multiple Inheritance

&gt; If a method exists in multiple methods Python looks from classes left to right to find the first matched method

**Usage**

```python

class ParentClass1()
    init;
    attributes;
    methods;

class ParentClass2()
    init;
    attributes;
    methods;


class childClass(ParentClass1,ParentClass2)
    init;

```

````python
#Example

class ClassicPhone:
    'Classic phones are old school phones, which can only be used for phone calls'
    def __init__(self,make,model):
      self.make = make
      self.model = model

    def makePhoneCall(self):
      print('Calling')

    def receivePhoneCall(self):
      print('Receiving')

class Camera:
    'Cameras are used to take pictures'
    def __init__(self,make,model):
      self.make = make
      self.model = model

    def takePicture(self):
      print('Taking Picture')

class CameraPhone(ClassicPhone,Camera):
    'New generation phones with inbuilt camera'
    def __init__(self,make,model):
      self.make = make
      self.model = model

#Create a new iphone 6 object from CameraPhone class
iphone6 = CameraPhone('Apple', 'iphone')

#Call MakePhoneCall method
iphone6.makePhoneCall()

#Call Receive Phone Call method
iphone6.receivePhoneCall()

#Call Take Picture Method
iphone6.takePicture()
````

# Polymorphism

&gt; Polymorphism in OOP is the ability to refer to multiple objects by a single generic object

&gt; The generic object is usually the parent class which can be used to refer to any of the objects created by the child class

````python
#Example

#Create an Animal Class
class Animal():
    'This is a Parent class for creating Animals'

    def __init__(self,name,breed,color):
        self.name = name
        self.breed = breed
        self.color = color


    def walk(self):
        print(self.name + ' is walking')

    def eat(self):
        print(self.name + ' is eating')

    def makeNoise(self):
        print(self.name + ' is making noise')

class Dog(Animal):
    'This is the Dog class inherited from Animal'

    # Override the makeNoise method to make the dog bark
    def makeNoise(self):
        print(self.name + ' is barking')


class Cat(Animal):
    'This is the cat class inherited from Animal'

    # Override the makeNoise method to make the cat meow
    def makeNoise(self):
        print(self.name + ' is meowing')


# Create a new Dog object dog1
dog1 = Dog('max', 'bulldog', 'black')

#Create a new Cat object cat1
cat1 = Cat('manny','tabby','orange')

#Adding dog1 and cat1 to a list names animals
animals = [dog1,cat1]

#Iterating over the list of animals and calling the makeNoise method
for animal in animals:
    animal.makeNoise()
````
